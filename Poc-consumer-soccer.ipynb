```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
req = requests.get('https://fbref.com/en/comps/24/stats/Serie-A-Stats')
if req.status_code == 200:
    print('Requisição bem sucedida!')
    content = req.content
```

```python
soup = BeautifulSoup(content, 'html.parser')
table = soup.find_all(name='table')  # Consulta varios elementos com a tag table
# table = soup.find_all(name='table', attrs={'id':'stats_player'}) # Consulta varios elementos com a tag table e id:
```

```python
# print(table)
table_str = str(table)
df = pd.read_html(table_str)[0]
# df[numeric_cols] = df[numeric_cols].apply(pd.to_numeric)
```
